Typowy przepływ sterowania w programie wykorzystuje np. takie struktury:

- jeżeli warunek $X$ jest spełniony, to wykonaj zadanie `A`, w przeciwnym przypadku wykonaj zadanie `B`
- jeżeli warunek $X$ jest spełniony, to wykonaj zadanie `A`, w przeciwnym przypadku jeżeli warunek $Y$ jest spełniony, to wykonaj zadanie `B`, w pozostałych przypadkach wykonaj zadanie `C`
- powtarzaj zadanie `A` $n$-krotnie
- wykonuj zadanie `B` tak długo, dopóki warunek $X$ jest spełniony

Każdy _imperatywny_ język programowania definiuje odpowiednie instrukcje umożliwiające opisanie powyższych zachowań (tzw. _control flow_).

### Przykład przepływu sterowania w _pseudo kodzie_ i w formie schematu blokowego

W zależności od sprzyjających warunków otwiera okno, w przeciwnym przypadku zamyka.

```
if opad:  # jeżeli pada deszcz, zamknij okno
```

````python
a = True
b = False
print( a or b )      # True
print( a and b )     # False
print( a and not b ) # True
````

### Operatory relacyjne

Operatory relacyjne zwracają, dla podanych argumentów, wynik `True` lub `False`.

Głównymi operatorami relacyjnymi są operatory porównania: `<`, `<=`, `>`, `>=` oraz `==` i `!=`. Ten ostatni oznacza _nie równe_.

W poniższym przykładzie sprawdzamy, czy `a` jest mniejsze czy większe od `b`:

````python
a = 10.0
b = 9.9
print(a < b)
print(a > b)
print(a == b)
c = "aaa"
d = "bbb"
print( c < d ) # porównanie leksykograficzne
````

Równość jest weryfikowana za pomocą operatora `==`, zaś operator `!=` jest
wykorzystywany do sprawdzenia, czy dwie zmienne nie są równe. Poniżej przykład wyjaśniający.

````python
# input("tekst zachety") - oznacza interaktywne czytanie danych z klawiatury
# int() - wykonuje konwersję tekstu na liczbę całkowitą
a = int( input("Podaj a: "))
b = int( input("Podaj b: "))
c = int( input("Podaj c: "))

print("Czy a jest równe b?", a == b)
print("Czy a jest równe c?", a == c)
print("Czy a nie jest równe c?", a != c)
print("Czy a jest mniejsze lub równe od b?", a <= b)
print("Czy a jest mniejsze lub równe od c?", a <= c)

# Sprawdzamy, czy kolory są takie same
kolor0 = 'blue'
kolor1 = 'green'
print("czy kolor0 jest taki sam jak kolor1?", kolor0 == kolor1)
````

### Wyrażenia logiczne

Wyrażenia logiczne są mniej lub bardziej skomplikowanymi fragmentami kodu, zwracającymi `True` lub `False`. Budując je, korzystamy z operatorów logicznych `and`, `or`, `not`, jak i operatorów relacyjnych.

````python
# Jeżeli 10 < 9 (fałsz) i 15 < 20 (prawda) -> fałsz
print(10 < 9 and 15 < 20)
````

````python
# Jeżeli 10 < 9 (fałsz) lub 15 < 20 (prawda) -> prawda
print(10 < 9 or 15 < 20)
````

Poniższy przykład wypisuje

- `True` jeśli jest przerwa obiadowa, i
- `True` jeśli jest po pracy.

````python
time = float( input("Podaj godzinę (zapisz z kropką, np. 13.05): ") )  # Aktualny czas

work_starts = 8.00  # Godzina rozpoczęcia dnia pracy
work_ends = 17.00   # Godzina zakończenia dnia pracy

lunch_starts = 13.00  # Godzina rozpoczęcia przerwy obiadowej
lunch_ends = 14.00    # Godzina zakończenia przerwy obiadowej

# Sprawdzenie, czy jest przerwa obiadowa
print("Czy mamy przerwę obiadową?")
is_lunchtime = time >= lunch_starts and time < lunch_ends
print(is_lunchtime)

# Sprawdzenie, czy jesteśmy poza godzinami pracy
print("Czy już po pracy?")
outside_working_hours = time < work_starts or time >= work_ends
print(outside_working_hours)
````

W przypadku bardziej skomplikowanych wyrażeń używamy nawiasów, aby określić kolejność ewaluacji.
W przykładzie podano warunek sprawdzający, czy podana liczba jest z przedziału $(10, 50) \cup (100, 200)$:

````python
value = 150.5
print ((value > 100 and value < 200) or (value > 10 and value < 50))
````

Dwa sprawdzenia w nawiasach są wykonywane najpierw (wyniki odpowiednio to: `True` oraz `False`), i dopiero wtedy zadziała operator '`or`', sprawdzający, czy choć jeden z warunków jest prawdziwy.

Powyższy kod można zapisać również w bardziej naturalny, dla matematyki, sposób:

````python
value = 150.5
print ( (100 < value < 200) or (10 < value < 50) )
````

Operatory `and` oraz `or` są _short-circuit evaluated_. Oznacza to, że wyrażenie jest oceniane od lewej do prawej strony tak długo, aż nie zostanie określony jego wynik. W chwili ustalenia wyniku dalsza ewaluacja jest zatrzymywana.

Przykład:

Rozważmy wyrażenie `(x > 0) and (math.sqrt(x) < 1)`.

* dla `x = 4` lewa część `(x > 0)` jest oceniona na `True` i ewaluacja jest kontynuowana dla prawej części `(math.sqrt(x) < 1)`. Ta zaś oceniona jest na `False`, więc całość wyrażenia logicznego jest oceniona na `False`
* dla `x = 1/4` lewa część `(x > 0)` jest oceniona na `True` i ewaluacja jest kontynuowana dla prawej części `(math.sqrt(x) < 1)`. Ta zaś oceniona jest na `True`, więc całość wyrażenia logicznego jest oceniona na `True`
* dla `x = -2` lewa część `(x > 0)` jest oceniona na `False`. Ponieważ wyrażenie z `and` jest prawdziwe tylko wtedy, gdy lewa i prawa strona są prawdziwe, całe wyrażenie jest ocenione na `False` bez sprawdzania drugiej część (_short-circuit evaluation_)

#### Przykład: przelicznik walut

Kantor wymienia funty angielskie na dolary amerykańskie w cenie niższej niż cena rynkowa. Wskaźniki obniżki ceny (mnożniki) podane są poniżej i zależą od ilości kupowanej waluty.

| Środki (GBP)                  | wskaźnik obniżki        |
|-------------------------------|-------------------------|
| Mniej niż $100$               | 0.9                     |   
| Od $100$ do $1 000$           | 0.925                   |   
| Od $1 000$ do $10 000$        | 0.95                    |   
| Od $10 000$ do $100 000$      | 0.97                    |   
| Powyżej $100 000$             | 0.98                    |   

Jednakże kantor pobiera prowizję za transakcję elektroniczną, doliczając 10% marży po wykonaniu wymiany przelewem na konto.

Jeśli transakcja jest gotówkowa, nie pobiera marży.

Aktualny przelicznik: 1 GBP jest równy 1.33153 USD.

Napisz program wykonujacy wymianę zadanej kwoty GBP

````python
GBP  = 15600.05  # Środki w GBP wymieniane na USD
cash = True  # True jesli transakcja gotówkowa, jesli elektroniczna, to False

market_rate = 1.33153  # przelicznik 1 GBP na dolary

# Logika wymiany, zgodna z tabelą
if GBP < 100:
    USD = 0.9*market_rate*GBP
elif GBP < 1000:
    USD = 0.925*market_rate*GBP
elif GBP < 10000:
    USD = 0.95*market_rate*GBP
elif GBP < 100000:
    USD = 0.97*market_rate*GBP
else:
    USD = 0.98*market_rate*GBP

if not cash:
    USD *= 0.9  # zapis jest uproszczeniem zapisu: USD = 0.9*USD

print("Środki w GBP do sprzedaży:", GBP )
print("Środki w USD zakupione:", round(USD, 2) )
print("Wskaźnik:", round( USD/GBP, 4) )
````

### Pętle

Pętle są istotą obliczeń wykonywanych za pomocą komputera. Dzieki jego szybkości pracy, możemy nakazać mu wielokrotne wykonanie tych samych czynności - zapisując to działanie za pomocą kilku linijek kodu.

#### Pętla `for` - z licznikiem

Petla `for` jest blokiem sterujacym, opisującym wykonanie operacji określoną liczbę razy. Nazywana jest pętlą z licznikiem. Często sotosowana do przetwarzania tablic (list elementów).

````python
for n in range(4):
    print("----")
    print(n, n**2)
````

Powyższa pętla wykonuje się 4 razy, kolejno dla wartości licznika `n`: `0`, `1`, `2` oraz `3`. Instrukcja

```python
for n in range(4):
```
mówi, że chcemy wykonać pętlę dla czterech liczb całkowitych, startując od `0` (patrz: https://docs.python.org/3/library/stdtypes.html#range - dokumentacja `range`).

Wartość `n` jest zwiększana o `1` po każdym wykonaniu pętli (iteracji). Kod, który powinien być wykonany w obrębie pętli zapisany **musi być** z wcięciem (tabulator lub co najmniej 2 spacje), ale z takim samym wcięciem:

```python
....print("----")
....print(n, n**2)
```

Pętla startuje od `0` i nie wykonuje się dla `n == 4`. Zapis `range(4)` jest skrótowym zapisem `range(0, 4)`, który oznacza: zaczynamy od `0`, zwiększamy licznik kolejno o `1` i **nie dochodzimy** do `4`.

Proces iteracji możemy również zapisać tak:

````python
for i in range(-2, 3):
    print(i, end=", ")
````

Pętla startuje od `-2`, ale nie dochodzi do `3`.

W przebiegu pętli możesz również podać krok z jakim wartość licznika będzie zwiększana (lub zmniejszana, jeśli ujemne):

````python
for n in range(0, 10, 3):
    print(n, end="; ")
````

#### Pętla `for` - dla sekwencji

Pętla `for` może być wykorzystana do przeglądania sekwencji danych (lub kolekcji danych). W takiej stytuacji wykonuje się tak długo, aż wszystkie elementy zostaną przeglądnięte (pętla się _wyczerpie_).

**Przykład**

Skrypt przegląda napis, litera po literze od początku, wypisując tylko literę `a` dużą lub małą, zaś pozostałe zastępuje kropką.

````python
napis = "Ala ma kota, As to Ali pies"
for litera in napis:
  if litera == "a" or litera == "A":
    print(litera, end="")
  else:
    print(".", end="")
````

**Przykład**

Skrypt przegląda listę ocen, oblicza pomocniczo sumę (`suma`) oraz liczbę ocen (`licznik`) i na końcu wypisuje średnią ocenę zaokrągloną do 2 miejsc po przecinku.

````python
lista = [2, 3, 2, 3.5, 3, 2, 4, 5, 2, 4.5, 5]
licznik = 0
suma = 0
for ocena in lista:
  suma += ocena
  licznik += 1

srednia = suma / licznik
print(f"srednia ocen: {srednia}")
print(f"srednia ocen: {srednia:.2f}")
````

#### Pętla `while` - z warunkiem logicznym

O ile pętla `for` wykonuje swoje zadanie określoną liczbę razy (określoną przez `range` lub długością sekwencji), o tyle pętla `while` powtarza zadanie, dopóki warunek logiczny jest spełniony. Na przykład:

````python
print("Start pętli while")
x = -2
while x < 5:
  print(x, end=" ")
  x += 1  # zwiekszamy x o 1 - inkerementacja
print()  # stawiamy "enter"
print("Koniec pętli while")
````

Treść petli `while`, podana po słowie kluczowym `while`, zapisana jest z użyciem 2 spacji wcięcia i wykonywana jest tak długo, dopóki `x < 5` jest wartościowane na `False`.

UWAGA: bardzo łatwo zawiesić komputer, wadliwie konstruując tę pętlę:

```python

  print(x)
```

Ta pętla wykonywać się będzie w nieskończoność, ponieważ zawsze `x < 5 == False`. Błąd polega na tym, że wewnątrz pętli `x` nigdy się nie zmienia. Jeśli chcesz spróbować, odkomentuj drugą i trzecią linię i uruchom kod.

Poniższy przykład pokazuje, jak maleją wartości `x` podnoszone kolejno do kwadratu, o ile $x < 1$.

````python
x = 0.9
while x > 0.001:
  # Kwadrat x (można to zapisać inaczej: x *= x albo x = x**2)
  x = x*x
  print(x)
````

Pętla zatrzymuje się, gdy `x <= 0.001`.

Jeśli jednak $x \ge 1$, wtedy pętla wpada w nieskończone wykonanie. Nie próbuj tego robić - albo nie, spróbuj zmienić `x = 1.9`. Gdy uruchomisz kod, zobaczysz, że pętla działać będzie w nieskończoność. W takiej sytuacji **będziesz musiał** zresetować kernel i przywrócić poprzednią wartość ` x = 0.9`.

Aby uczynić kod bezpieczniejszym, powinniśmy wstawić sprawdzenie, czy $x < 1$, zanim wejdziemy do pętli `while`.

````python
x = 1.9
if x < 1:
  while x > 0.001:
    # Kwadrat x (można to zapisać inaczej: x *= x)
    x = x*x
    print(x)
else:
  print("x nie jest mniejszy niż 1!")
````

##### Instrukcja `continue`

Czasami, wykonując pętlę, chcemy przerwać jej bieżące wykonanie i przejść do kolejnego kroku pętli.
W takiej sytuacji użyjemy słowa kluczowego `continue`.
Poniżej przykład, demonstrujący petlę wykonującą się 20 razy (od `j = 0` do `19`), w której w każdym kroku sprawdzamy, czy liczba `j` jest podzielna przez `4`. Jeśli nie, to drukujemy stosowny komunikat i przechodzimy do kolejnego kroku ("obrotu") pętli. Jeśli jest podzielna przez `4`, przeskakujemy do kolejnego kroku, bez wypisywania komunikatu.

````python
for j in range(20): # od 0 do 19
    if j % 4 == 0:  # Sprawdzamy podzielność przez 4, za pomocą modulo
        continue    # przeskakujemy do kolejnej iteracji względem j
    print("Liczba nie podzielna przez 4:", j)
````

#### Pętle nieskończone

Powszechnym błędem, szczególnie przy użyciu pętli `while`, jest wprowadzenie obliczeń w nieskończony cykl ([infinite loop](https://en.wikipedia.org/wiki/Infinite_loop)).

Pętle nieskończone wprowadzają komputer w stan niestabilności. Czasami wymagane będzie jego zresetowanie.

Dobrą praktyką jest zabezpieczanie się przed nieskończonymi obliczeniami:

````python
x = 0.0

licznik = 0
while x < 0.05:

    # obrona przed pętlą nieskończoną
    licznik += 1
    if licznik > 2000:
        print("Petla przekroczyła 2000. Wychodzę z niej!")
        break
````

Jednak niektórzy programiści świadomie używają pętli nieskończonej, zapewniając wyjście z niej odpowiednim warunkiem w jej treści:

````python
x = 1
while True:
  print(x)
  if x == 5:
    break
  x += 1

print("Koniec obliczeń")
````

#### Pętle `for` oraz `while` z warunkeim `else`

W języku Python pętle `for` oraz `while` mają rozszerzoną formę - z warunkiem `else`. Kod po tym warunku wykona się, jeśli wyjście z pętli odbywa się z powodu wyczerpania przeglądanych wartości i nie zostaje przerwane przez `break`.

````python
# uruchom dwa razy,
# 1. podając napis ze znakiem `-` w środku napisu
# 2. nie podając znaku `-` w napisie
napis = input("podaj ciąg znaków : ")
for znak in napis:
  if znak == "-":
    print("\n pojawił się znak `-`. Przerywam")
    break
  print(znak, end="")
else:
  print("\n nie było znaku `-`. Osiągnięto koniec napisu")

print("\n pętla się skończyła")
````

### Zadanie 4 - najwięszy wspólny podzielnik

Napisz program obliczający najwiekszy wspólny podzielnik dwóch liczb całkowitych `a` i `b` podanych przez użytkownika. Przetestuj program dla dużych liczb.

Problem ten rozwiążesz za pomocą algorytmu Euklidesa - rozwiązanie w pseudokodzie znajdziesz na Wikipedii.

````python
def nwd(a, b):
    while b:
        a, b = b, a % b
    return a

a = int(input("Podaj pierwszą liczbę: "))
b = int(input("Podaj drugą liczbę: "))

print("Największy wspólny podzielnik dla liczb", a, "i", b, "to", nwd(a,b))

````
